```python
# 1. printing the game board
# 2. take player input
# 3. check for win or tie
# 4. switch the player
# 5. check for win or tie again.

# importing  the random module for computer (oponent)
# this is imported at the end when we are making the computer oponent

import random

board = ["-","-","-",
         "-","-","-",
         "-","-","-"]
currentPlayer= "X"
winner= None
gameRunning = True
# 1. printing the game board :-

def printBoard(board):
    print(board[0]+" | "+board[1]+" | "+board[2])
    print("----------")
    print(board[3]+" | "+board[4]+" | "+board[5])
    print("----------")
    print(board[6]+" | "+board[7]+" | "+board[8])

# 2. take player input :-

def playerInput(board):
    inp= int(input("Enter a number 1-9 :-"))
    if inp>=1 and inp<=9 and board[inp-1] == "-":
        board[inp-1] = currentPlayer
    else:
        print("Oh! player is already been in that spot.")

# 3. check for win or tie
# Check the horizontal board:-

def checkHorizontal(board):
    global winner
    if board[0]==board[1]==board[2] and board[1] != "-":
        winner = board[1]
        return True
    elif board[3]==board[4]==board[5] and board[4] != "-":
        winner = board[4]
        return True
    elif board[6]==board[7]==board[8] and board[7] != "-":
        winner = board[7]
        return True

# Checking the vertical board:-

def checkRow(board):
    global winner
    if board[0] == board[3] == board[6] and board[3] != "-":
        winner = board[3]
        return True
    elif board[1] == board[4] == board[7] and board[1] != "-":
        winner = board[1]
        return True
    elif board[2] == board[5] == board[8] and board[5] != "-":
        winner = board[5]
        return True

# Checking Diagonal

def checkDiag(board):
    global winner
    if board[0]==board[4] == board[8] and board[0] != "-":
        winner = board[0]
        return True
    if board[2]==board[4] == board[6] and board[2] != "-":
        winner = board[2]
        return True

# Checking if there is a tie :-
def checkTie(board):
    global gameRunning
    if "-" not in board:
        printBoard(board)
        print("It is a tie! ")
        gameRunning = False

# Checking for win:-

def checkWin():
    if checkDiag(board) or checkHorizontal(board) or checkRow(board):
        print(f"The winner is {winner}")
        gameRunning = False



# 4. switch the player

def switchPlayer():
    global currentPlayer
    if currentPlayer == "X":
        currentPlayer = "O"
    else:
        currentPlayer = "X"

# Computer (oponent) :-
def computer(board):
    while currentPlayer =="O":
        position = random.randint(0,8)
        if board[position] == "-":
            board[position] = "O"
            switchPlayer()

# 5. check for win or tie again.

while gameRunning:
    printBoard(board)
    playerInput(board)
    checkWin()
    checkTie(board)
    switchPlayer()
    computer(board)
    checkWin()
```
